```python
number_rating = ratings_with_books.groupby('Title')['Book-Rating'].count().reset_index()
```

```python
number_rating.rename(columns={'Book-Rating':'num_of_rating'},inplace=True)
```

```python
final_rating = ratings_with_books.merge(number_rating, on='Title')
```

```python
final_rating = final_rating[final_rating['num_of_rating'] >= 50]
```

```python

final_rating.head()
```

```python
# lets drop the duplicates
final_rating.drop_duplicates(['User-ID','Title'],inplace=True)
```

```python

final_rating.isnull().sum()
```

```python
# Lets create a pivot table
book_pivot = final_rating.pivot_table(columns='User-ID', index='ISBN', values= 'Book-Rating')
```

```python
book_pivot
```

```python

book_pivot.shape
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
book_pivot.iloc[3,:]
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(n_neighbors=6, metric='cosine')
```

```python
model.fit(book_sparse)
```

```python
isbn_to_title = dict(zip(final_rating['ISBN'], final_rating['Title']))
title_to_isbn = dict(zip( final_rating['Title'],final_rating['ISBN']))
```

```python
def recommend_books(book_name):
    isbn=title_to_isbn[book_name]
    book_index = np.where(book_pivot.index == isbn)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_index, :].values.reshape(1, -1))

    print(f"Because you liked {isbn_to_title[isbn]}, you might also like:\n")
    for i in range(1, len(suggestion[0])):  # Start from 1 to skip the same book
        print(isbn_to_title.get(book_pivot.index[suggestion[0][i]]))
    
    return suggestion
```

```python
suggestion = recommend_books("The Secret Garden")
suggestion
```
